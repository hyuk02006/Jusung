### 교통사고 데이타셋에서 요일별 사망자수 필터링 + 시각화

(1) 요일별 사망자수 필터링 (groupby() 이용 )

(2) 요일별 사망자수의 합계 (sum() 추가)

(3) 그래프

```python
# 1. 요일별 중대교통사고 사상자 합계 ( 사상자 10명 이상 )

%matplotlib notebook

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

# 한글깨짐처리
font_loc = "c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_loc).get_name()
rc('font', family=font_name)
```

```python
# 데이타 읽어오기
data=pd.read_csv('data/accidentdata.csv')
data
```

```python
# 사상자수가 10이상인 데이타만 필터링
df = data[data['사상자수']>=10]
df
```

[ 연습 ] 요일별 사망자수를 그래프로 표현

```python
# 요일별로 사상자수 합계
df_days = df.groupby('요일')
df_days
```

```python
# result = df_days['사상자수'].sum()  # 요일별 그룹한 후 사상자수의 합계
# result
```

```python
df_days['사상자수'].sum().plot(kind='bar', title='요일별 사상사 분석')
```

```python
# 요일 글씨가 눕혀져 있기에
replot = df_days['사상자수'].sum().plot(kind='bar', title='요일별 사상사 분석')
replot.set_xticklabels(['금요일','목요일','수요일','월요일','일요일','토요일','화요일'],
                      fontsize=10, rotation=0)
```
