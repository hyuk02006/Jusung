추가 그래프 확인

```python
import seaborn as sns                
import matplotlib.pyplot as plt  

tips = sns.load_dataset('tips')
```

```python
""" (1) 시간대별로 지불금액의 평균을 나타내는 바이올린 그래프에 성별을 추가한다면 """
plt.subplot()
# axes = sns.violinplot(x='time', y='total_bill', data=tips )
axes = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex')
# axes = sns.violinplot(x='time', y='total_bill', data=tips, hue='sex', split=True)
```

```python
""" (2) 지불금액과 팁에 대한 산점도 그래프에 성별을 추가한다면 """

# axes = sns.regplot(x='total_bill', y='tip', data=tips , fit_reg=False) 
# regplot()에는 hue 속성이 없기에 lmplot() 이용 (소문자 엘엠플롯)
# fit_reg=True 선을 그려준다
# axes = sns.lmplot(x='total_bill', y='tip', data=tips , fit_reg=False)
axes = sns.lmplot(x='total_bill', y='tip', data=tips , fit_reg=False, hue='sex')

# 점의 마커를 변경
# axes = sns.lmplot(x='total_bill', y='tip', data=tips , fit_reg=False, hue='sex', markers=['o','x'])

# 점의 크기 변경 - 인원수에 따라
# axes = sns.lmplot(x='total_bill', y='tip', data=tips , fit_reg=False, hue='sex', scatter_kws={'s':tips['size']*10})
```

```python
""" FacetGrid()를 이용한 그룹별 그래프 """
# 시간대별로 지불 금액을 표현한 그래프
facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill')

# 요일별로 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='day', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
facet.add_legend()  # 성별 글씨 보이게

# 시간대별 흡연여부별 지불 금액을 표현한 그래프에 성별 구분 추가
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
facet.add_legend() 
```
