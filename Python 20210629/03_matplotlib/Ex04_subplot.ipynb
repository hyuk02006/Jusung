## 하나의 화면에 여러 개의 그래프 그려보기

```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig=plt.figure()
ax1 = fig.add_subplot(1,2,1) # 1행 2열의 공간에서 첫번째
ax2 = fig.add_subplot(1,2,2) # 1행 2열의 공간에서 두번째
ax1.plot( np.random.randn(100))
ax2.plot( np.random.randn(200).cumsum());
```

```python
# x, y 축 값 범위 지정 
fig=plt.figure()
ax=fig.add_subplot(111) # (1,1,1) 동일
ax.plot(np.random.randn(50),'b*--');   # b:blue --:선모양  *:꼭지모양
ax.set_xlim([1,100])
ax.set_ylim([-5,5])
# ax.set_title('MY Graph') # 한글깨짐
# ax.set_xlabel('data_x')
```

## 그래프 속성 지정

- title() : 
- xlabel() : 
- ylabel() : 
- gri() : 격자 눈금
- text() : 글씨를 출력 (tooㅣtip을 간단하게 하고자 할 때)

```python
# plt에서 그래프 속성 지정
%matplotlib inline
plt.plot(np.random.randn(10));
plt.xlabel('data_x')
plt.ylabel('value_y')
plt.title('Chart')
plt.text(0.5, 0.5,'Random Value') # x축과 y축은 0과 1 사이
plt.grid(True)
plt.show()
```

### x 축 간격

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.plot(np.random.randn(40), 'r-+') # 'r-+'
ax.set_xticks([10,20,30,40, 50])    # tick : 간격( 랜덤 40개까지 출력됨 )
ax.set_xticklabels(['one','two','three','four','five'])
```

```python
# figure로 전체 그래프의 액자 같은 개념으로 액자 크기 조절이 가능하다
fig=plt.figure(figsize=(8,5))
plt.plot(np.random.randn(40), 'r-+')
```
