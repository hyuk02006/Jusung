```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

""" 1.  figure(그림틀=공간?)을 만들고 거기에 subplot을 추가 하는 방식 """
fig=plt.figure()
ax1 = fig.add_subplot(1,2,1) # 1행 2열의 공간에서  첫번째
ax2 = fig.add_subplot(1,2,2) # 1행 2열의 공간에서  두번째
ax1.plot( np.random.randn(100))
ax2.plot( np.random.randn(200).cumsum())
```

```python
""" 2. figure로 공간을 만들고 파이플롯(pyplot)의 subplot()으로 각각의 패널 지정 """
fig=plt.figure()
ax1 = plt.subplot(1,2,1) # 1행 2열의 공간에서 첫번째
ax2 = plt.subplot(1,2,2) # 1행 2열의 공간에서 두번째
ax1.plot( np.random.randn(100))
ax2.plot( np.random.randn(200).cumsum())
```

```python
""" 3. 간단하게 ax 객체를 이용하기 """
fig, ax = plt.subplots(1,2)        # 리턴값이 2개이다.(figure 와 axes ) -> ax를 받기 위해서 앞에 f도 받아야 한다 
ax[0].plot( np.random.randn(100))
ax[1].plot( np.random.randn(200).cumsum())
```

```python
''' 4. 그래프 전체 크기를 지정 '''
fig, ax = plt.subplots(1,2, figsize=(16,8))       
ax[0].plot( np.random.randn(100));
ax[1].plot( np.random.randn(200).cumsum());
```
