## datetime 라이브러리 

- date 
- time 
- datetime
    

   > [  df  ]   
 문자열 처리를 위해 str 접근자를 사용하듯이 datetime 오브젝트는 dt 접근자 사용해야 
 datetime 속성이나 메소드를 사용가능
 
     [도서] Do it 데이타분석을 위한 판다스 입문

```python
""" datetime 라이브러리 : date / time / datetime 오브젝트 """

from datetime import datetime

# 현재 시간 출력
print(datetime.now())
print(datetime.today())

# 시간 계산
t1 = datetime.now()
t2 = datetime(2000, 1, 1)
print(t1-t2)

# 시간 형식 지정자로 자세히 출력
now = datetime.now()
print(now.strftime('%Y-%m-%d'))  # 년월일
print(now.strftime('%H:%M:%S'))  # 시분초

print(now.year)
print(now.month)
print(now.day)
```

## [ 연습 ]  datatime  오브젝트로 변경하여 필요한 데이타 추출 및 계산

```python
import pandas as pd


ebola = pd.read_csv('./data/ebola_timeseries.csv')
ebola.head()
```

```python
# datetime 오브젝트로 변경하여 필요한 데이타 추출 및 계산

print(ebola.info())  # [결과확인] Date 122 non-null object 형임
ebola['date_dt'] = pd.to_datetime(ebola['Date'])
print('-'*50)
print(ebola.info())  # [추가된 컬럼] date_dt 122 non-null datetime64[ns] 형임
```

```python
# parse_dates 속성으로 datetime 오브젝트로 바로 변환
ebola = pd.read_csv('./data/ebola_timeseries.csv', parse_dates=['Date'])
print(ebola.info())  # [결과확인] Date  122 non-null datetime64[ns]

# 확인후 다시 윗부분 실행 (date_dt 컬럼 필요)
```

```python
# Date와 date_dt 오브젝트 출력
# print(ebola[['Date','date_dt']])

# 날짜에서 년, 월, 일을  추출하여 새로운 컬럼 생성
# ebola['year'] = ebola['date_dt'].year   # 에러
"""
 문자열 처리를 위해 str 접근자를 사용하듯이 datetime 오브젝트는 dt 접근자 사용해야 
 datetime 속성이나 메소드를 사용가능
"""
ebola['year'] = ebola['date_dt'].dt.year
print(ebola[['Date','date_dt','year']])
```
